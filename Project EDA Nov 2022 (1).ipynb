```python
#-6-Features to be check for outliers
feature_names = ['Auction_Avg_price','Auction_Clean_price','Retail_Avg_price']
##Calculate Q1, Q3 and IQR for each feature
Q1=[]
Q3=[]
IQR=[]
for i,j in enumerate(feature_names):
    Q1.append(df_new[j].quantile(0.25))
    Q3.append(df_new[j].quantile(0.75))
    IQR.append(Q3[i]-Q1[i])
#Dropping outliers
for i, j in enumerate(feature_names):
    true_index = (df_new[j] >= (Q1[i] - 2 * IQR[i])) & (df_new[j] <=(Q3[i] + 2 * IQR[i]))
    df_new = df_new.loc[true_index] 
```
